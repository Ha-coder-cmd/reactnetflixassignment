```python
import pandas as pd
from collections import Counter 
```

```python
df=pd.read_csv(r"C:\Users\ALL\Downloads\netflix_titles.csv")
```

```python
df
```

```python
df['cast']
```

```python
def extract_actors(castlist):
    if pd.notna(castlist):
        actors=castlist.split(", ")
        return actors 
    return []
```

```python
all_actors=df['cast'].apply(extract_actors).explode().dropna().tolist()
```

```python
all_actors
```

```python
actor_count=Counter(all_actors)
```

```python
actor_count
```

```python
actor_df=pd.DataFrame(actor_count.items(),columns=["Actor","Frequency"])
```

```python
actor_df
```

```python
actor_df=actor_df.sort_values(by=["Frequency"],ascending=False)
```
